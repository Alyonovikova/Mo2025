#### 12. Если вы обнаружили пропуски в данных, то заполните их на основе среднего или медианы или моды

```python
df['price'] = df['price'].replace(np.nan, df['price'].mode()[0])
df['peak-rpm'] = df['peak-rpm'].replace(np.nan, df['peak-rpm'].mode()[0])
df['horsepower'] = df['horsepower'].replace(np.nan, df['horsepower'].mode()[0])
df['stroke'] = df['stroke'].replace(np.nan, df['stroke'].mode()[0])
df['bore'] = df['bore'].replace(np.nan, df['bore'].mode()[0])
df['num-of-doors'] = df['num-of-doors'].replace(np.nan, df['num-of-doors'].mode()[0])
```

```python
df['price'] = df['price'].fillna(df['price'].median())
df['peak-rpm'] = df['peak-rpm'].fillna(df['peak-rpm'].median())
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].median())
df['stroke'] = df['stroke'].fillna(df['stroke'].median())
df['bore'] = df['bore'].fillna(df['bore'].median())
df['num-of-doors'] = df['num-of-doors'].fillna(df['num-of-doors'].median())
```

#### 23. Преобразуйте все категориальные поля в числовые с помощью LabelEncoder

```python
data_cat=df.select_dtypes(exclude=np.number)
data_num=df.select_dtypes(np.number)
data_cat=data_cat.reset_index()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['make'] = encoder.fit_transform(df['make'])
df['make'] = df['make'].astype(np.number)

df['fuel-type'] = encoder.fit_transform(df['fuel-type'])
df['fuel-type'] = df['fuel-type'].astype(np.number)

df['aspiration'] = encoder.fit_transform(df['aspiration'])
df['aspiration'] = df['aspiration'].astype(np.number)

df['num-of-doors'] = encoder.fit_transform(df['num-of-doors'])
df['num-of-doors'] = df['num-of-doors'].astype(np.number)

df['body-style'] = encoder.fit_transform(df['body-style'])
df['body-style'] = df['body-style'].astype(np.number)

df['drive-wheels'] = encoder.fit_transform(df['drive-wheels'])
df['drive-wheels'] = df['drive-wheels'].astype(np.number)

df['engine-location'] = encoder.fit_transform(df['engine-location'])
df['engine-location'] = df['engine-location'].astype(np.number)

df['engine-type'] = encoder.fit_transform(df['engine-type'])
df['engine-type'] = df['engine-type'].astype(np.number)

df['num-of-cylinders'] = encoder.fit_transform(df['num-of-cylinders'])
df['num-of-cylinders'] = df['num-of-cylinders'].astype(np.number)

df['fuel-system'] = encoder.fit_transform(df['fuel-system'])
df['fuel-system'] = df['fuel-system'].astype(np.number)
```

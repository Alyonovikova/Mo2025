#### 19. Постройте гистограмму по полю stroke

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data_url = 'https://raw.githubusercontent.com/yakushinav/journ/refs/heads/main/datasets/automobile.csv'
df = pd.read_csv(data_url)
df = df.replace('?', np.nan)
df.columns = df.columns.str.replace('-', '_')
df.columns = df.columns.str.replace(' ', '_')
df['stroke'] = df['stroke'].astype(float)
df['stroke'] = df['stroke'].fillna(df['stroke'].mean())
plt.figure(figsize=(12, 8))
plt.hist(df['stroke'], bins=20, color='lightblue', edgecolor='black')
plt.title('Распределение хода поршня')
plt.xlabel('Ход поршня')
plt.ylabel('Частота')
plt.grid(True)
plt.show()
```
